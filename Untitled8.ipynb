```python
import numpy as np
input_size=2
hidden_size=2
output_size=1
weights_input_hidden=np.random.randn(input_size,hidden_size)
weights_input_hidden=np.random.randn(input_size,hidden_size)
biases_hidden=np.random.randn(1,hidden_size)
weights_hidden_output=np.random.randn(hidden_size,output_size)
bias_output=np.random.randn(1,output_size)
```

```python
input_values=np.array([[1.0,2.0]])
weighted_sum_hidden=np.dot(input_values,weights_input_hidden)+biases_hidden
weighted_sum_output=np.dot(weighted_sum_hidden,weights_hidden_output)+biases_hidden
print(weighted_sum_hidden)
print(weighted_sum_output)
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
activation_hidden_first_node = sigmoid(weighted_sum_hidden[:, 0])
print("Activation at the first node in the hidden layer:")
print(activation_hidden_first_node)
```

```python
activation_hidden_second_node = sigmoid(weighted_sum_hidden[:, 0])
print("Activation at the first node in the hidden layer:")
print(activation_hidden_second_node)
```

```python
activation_hidden=sigmoid(weighted_sum_hidden)
print("Activation on Hidden Layer:")
print(activation_hidden)
```

```python
weighted_sum_output=np.dot(activation_hidden,weights_hidden_output)+bias_output
print("Weighted sum at the output layer:")
print(weighted_sum_output)
```

```python
output_of_network=sigmoid(weighted_sum_output)
print("Output of the network:")
print(output_of_network)
```
